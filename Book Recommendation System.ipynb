# Popularity Based Recommendation System

```python
number_of_ratings = merge_df.groupby("Book-Title").count()["Book-Rating"].reset_index()
number_of_ratings.rename(columns={"Book-Rating": "Number of Ratings"}, inplace=True)
number_of_ratings
```

```python
number_of_ratings.sort_values("Number of Ratings",ascending=False)
```

```python
avg_ratings = merge_df.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_ratings.rename(columns={"Book-Rating": "Avg of Ratings"}, inplace=True)
avg_ratings
```

```python
popular_df=number_of_ratings.merge(avg_ratings,on="Book-Title")
popular_df
```

```python
popular_df = popular_df[popular_df["Number of Ratings"]>=200].sort_values("Avg of Ratings",ascending=False).head(100)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books_df,on="Book-Title").drop_duplicates("Book-Title")[['Book-Title','Book-Author','Image-URL-M','Number of Ratings','Avg of Ratings']]
```

```python
merge_df.groupby('User-ID').count()['Book-Rating']
```

```python
merge_df.groupby('User-ID').count()['Book-Rating']>200
```

```python
x = merge_df.groupby('User-ID').count()['Book-Rating']>200
x[x] ##boolean indexing
```

```python
active_users = x[x].index #fetching the user-Id
```

```python
active_users
```

```python
filtered_rating = merge_df[merge_df['User-ID'].isin(active_users)]
```

```python
filtered_rating
```

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50 ##ratings more than 50
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50 
famous_books = y[y].index
```

```python
famous_books
```

```python
filtered_rating['Book-Title'].isin(famous_books)
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates() ##no duplicates
```

```python
final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot_table = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot_table.fillna(0,inplace=True)
```

```python
pivot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pivot_table)
```

```python
similarity_scores
```

```python
similarity_scores.shape
```

```python
#fetching the index number
#for example- 
np.where(pivot_table.index=='Year of Wonders')[0][0] 
```

```python
similarity_scores[701] #books which are similar to 'Year of Wonders'
```

```python
list(enumerate(similarity_scores[701])) #gives distance as well as index
```

```python
sorted(list(enumerate(similarity_scores[701])),key=lambda x:x[1],reverse=True)
```

```python
sorted(list(enumerate(similarity_scores[701])),key=lambda x:x[1],reverse=True)[1:11] #top 10
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pivot_table.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pivot_table.index[i[0]])
```

```python
recommend("You Belong To Me")
```

```python
recommend("1st to Die: A Novel")
```

```python
similarity_data=pd.DataFrame(similarity_scores,index=pivot_table.index)
```

```python
def recommend_book(book_name):
    #index fetch
    index = np.where(similarity_data.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_data.iloc[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(similarity_data.index[i[0]])
```

```python
recommend_book("You Belong To Me")
```
